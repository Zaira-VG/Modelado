```
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```

def cap5_pendulo(t, x):
    # Vector de estados
    q = x[0]  # Posicin articular
    qp = x[1]  # Velocidad articular

    # Parmetros del pndulo
    m = 5  # Masa
    lc = 0.01  # Centro de masa
    g = 9.81  # Constante de aceleracin gravitacional
    b = 0.14  # Viscosidad
    fc = 0.45  # Coeficiente de friccin
    Ir = 0.16  # Momento de inercia del rotor
    tau = 1.5 * np.sin(t)  # Par aplicado

    # Aceleracin articular del pndulo
    qpp = (tau - b * qp + fc * np.tanh(100000 * qp) - m * g * lc * np.sin(q)) / Ir

    # Vector de salida
    xp = np.array([qp, qpp])  # xp[0]=x[1] velocidad articular, xp[1]=qpp aceleracin articular

    return xp



# Parmetros de simulacin
ti = 0  # Tiempo inicial
tf = 10  # Tiempo final de simulacin (segundos)
h = 0.0025  # Incremento de simulacin

# Funcin del pndulo
def cap5_pendulo(t, x):
    q, qp = x  # Posicin y velocidad articular
    m, lc, g, b, fc, Ir = 5, 0.01, 9.81, 0.17, 0.05, 0.16
    tau = 1.5 * np.sin(t)  # Par aplicado
    qpp = (tau - b * qp - fc * np.tanh(100000 * qp) - m * g * lc * np.sin(q)) / Ir
    return [qp, qpp]

# Simulacin usando solve_ivp
ts = np.arange(ti, tf, h)
sol = solve_ivp(cap5_pendulo, [ti, tf], [0, 0], t_eval=ts, method='RK45')

# Obtencin de coordenadas
theta = np.rad2deg (sol.y[0])
theta_dot= sol.y[0]

# Grfica de resultados
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(sol.t, theta)
plt.xlabel('Tiempo (s)')
plt.ylabel(' ()')
plt.title('Posicin articular')

plt.subplot(2, 1, 2)
plt.plot(sol.t, theta_dot, 'r')
plt.xlabel('Tiempo (s)')
plt.ylabel(' (/s)')
plt.title('Velocidad articular')

plt.tight_layout()
plt.show()
```
